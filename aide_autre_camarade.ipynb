```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer, make_column_selector
```

```python
iris = pd.read_csv("iris.csv")
iris['test'] = "a"
iris['test2'] = "b"
iris
```

```python
numerical_features = make_column_selector(dtype_include=np.number)
categorical_features = make_column_selector(dtype_exclude=np.number)

preprocessor = make_column_transformer((StandardScaler(), numerical_features),
                        (OneHotEncoder(), categorical_features))


preprocessor
```

```python
X = iris.drop(['class'], axis=1)
Y = iris['class']
preprocessor.fit_transform(iris)

iris_preprocessing = pd.DataFrame()

iris_preprocessing[preprocessor.get_feature_names_out()] = preprocessor.fit_transform(X)

iris_preprocessing
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_preprocessing,
                                                    Y,
                                                    test_size=0.33,
                                                    random_state=0,
                                                    stratify=iris["test"]) 
X_train = preprocessor.fit_transform(X_train)
```
